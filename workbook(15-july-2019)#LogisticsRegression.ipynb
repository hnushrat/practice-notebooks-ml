```python
#drops ROWS having null values
df.dropna(inplace=True)
```

```python
df.count()
```

```python
#Encoding: Conversion of categorical value to its numeric representation.
#Types: 1.Numeric encoding ; 2.One-hot Encoding
```

```python
sex=pd.get_dummies(df['sex'],drop_first=True)
sex.head(5)
#shows 0 and 1 ; if '0' then the column name doesn't contain itself else otherwise.
```

```python
embark=pd.get_dummies(df['embarked'],drop_first=True)
embark
```

```python
pcl=pd.get_dummies(df["pclass"],drop_first=True)
pcl
```

```python
df=pd.concat([df,sex,embark,pcl],axis=1)
```

```python
df
```

```python
df.drop(['pclass','sex','embarked','name','ticket'],axis=1,inplace=True)
```

```python
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
x=df.drop('survived',axis=1)
y=df['survived']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions=logmodel.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
res=confusion_matrix(y_test,predictions)
res
```
